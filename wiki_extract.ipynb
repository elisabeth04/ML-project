```python
result = wikipedia.search("Neural networks")  #get a random page content
page = wikipedia.page(result[0])
content = page.content
```

```python
with open("sample.txt", "w") as infile: #save the content
    infile.write(content)


```

```python
tokenizer = PunktSentenceTokenizer() #create sentence tokenization
tokenizer.train(content)
sentences= tokenizer.tokenize(content)
```
